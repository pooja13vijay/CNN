This algorithm uses 2 types of images namely, content images and style images to produce output. The style image has style of image to be transfered (ex, Van Gogh drawings) while the content image has content of images to be transfered (ex, image of a house). 

The content of the image is contained in the activations of different feature maps in layers in CNN while the style of the image is contained in correlations of feature aps in different layers in CNN. Representation of content and style of images are seperable. 

This algorithm makes use of pre-trained VGG-19 network. The higher layer in network capture high-level content in terms of object and their arrangement while the lower layers capture detailed pixel information.
